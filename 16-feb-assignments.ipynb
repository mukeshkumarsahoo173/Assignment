# Q5. Explain Primary Key and Foreign Key.

Ans


## Primary key:
* An identifier of a table which is unique and not null is called primary key
* A table can not have more than one primary key.
* Primary key can be defined at the column or the table level.
* If you create a composite primary key, it should be defined at the table level

## Foreign Key:


* A field as example c1 which is primary key for a table example T1 also the same c1 present in T2 now if T2 has a diffrent primery key as c2 than c1 is foreign key for table2.

* Foreign key is a column that creates a relationship between two tables.

* The purpose of the Foreign key is to maintain data integrity and allow navigation between two different instances of an entity. It acts as a cross-reference between two tables as it references the primary key of another table.

* Every relationship in the database should be supported by a foreign key.
